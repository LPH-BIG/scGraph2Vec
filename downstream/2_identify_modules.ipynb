```python
from sklearn.preprocessing import normalize
from sklearn.feature_selection import chi2,mutual_info_regression
import scipy.cluster.hierarchy as shc
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score,calinski_harabasz_score,davies_bouldin_score
```

```python
data11 = pd.read_csv("./data/tsne_perplexity40_run0_tsnekmeans10.csv")
data12 = pd.read_csv("./data/tsne_perplexity40_run1_tsnekmeans10.csv")
data13 = pd.read_csv("./data/tsne_perplexity40_run2_tsnekmeans10.csv")
data14 = pd.read_csv("./data/tsne_perplexity40_run3_tsnekmeans10.csv")
data15 = pd.read_csv("./data/tsne_perplexity40_run4_tsnekmeans10.csv")
data16 = pd.read_csv("./data/tsne_perplexity40_run5_tsnekmeans10.csv")
data17 = pd.read_csv("./data/tsne_perplexity40_run6_tsnekmeans10.csv")
data18 = pd.read_csv("./data/tsne_perplexity40_run7_tsnekmeans10.csv")
data19 = pd.read_csv("./data/tsne_perplexity40_run8_tsnekmeans10.csv")
data20 = pd.read_csv("./data/tsne_perplexity40_run9_tsnekmeans10.csv")
```

```python
link11=shc.linkage(data11.iloc[:,19:21], method='ward')
silhouette11 = []
k_sil11 = []
for i in range(2,150):
    fc=shc.fcluster(link11, t=i, criterion='maxclust')
    np.unique(fc,return_counts=True)
    score=silhouette_score(data11.iloc[:,19:21],fc)
    silhouette11.append(score)
    k_sil11.append(i)
```
